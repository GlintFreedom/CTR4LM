```python
lines = open('train.txt', 'r').readlines()
writer = open('ratings_final.txt', 'a', encoding='utf-8')
user_dict = dict()

for l in lines:
    tmp = l.strip()
    inter = [int(i) for i in tmp.split()]

    if len(inter) > 1:
        user_id, item_ids = inter[0], inter[1:]
        item_ids = list(set(item_ids))

        for item_id in item_ids:
            writer.write('%d\t%d\t1\n' % (user_id, item_id))
        unwatched_set = item - set(item_ids)
        for item_id in np.random.choice(list(unwatched_set), size=len(item_ids), replace=False):
            writer.write('%d\t%d\t0\n' % (user_id, item_id))
writer.close()

```

```python
kg = pd.read_table('kg_final.txt', sep = '\t', names = ['h', 'r', 't'])
kg
```

```python
entity = set(kg['h'].tolist()+ kg['t'].tolist())
len(entity)
```

```python
entity = list(entity)
entity
```

```python
writer = open('entityold2new.txt', 'w', encoding='utf-8')
entityold2new = dict()
for i in range(len(entity)):
    writer.write("%d\t%s\n" % (i, entity[i]))
    entityold2new[entity[i]] = i
entityold2new
```

```python
kg.h = kg.h.map(lambda x: entityold2new[x])
kg.t = kg.t.map(lambda x: entityold2new[x])
```

```python
kg.to_csv('kg_final.txt', sep='\t',index=False, header = False)
```
